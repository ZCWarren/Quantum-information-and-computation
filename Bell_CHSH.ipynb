# 1. Create Bell states

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qc = QuantumRegister(2, 'q')
circuit = QuantumCircuit(qc)

circuit.x(qc[0])
circuit.x(qc[1])
circuit.h(qc[0])
circuit.cx(qc[0], qc[1])
circuit.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
sv = Statevector.from_label('00')
new_sv = sv.evolve(circuit)
print(new_sv)
plot_state_qsphere(new_sv.data)
```

```python
counts = new_sv.sample_counts(shots=1000)

from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

# 2. Measurement

2.1 Measure on $Z$ axis and $(Z+X)/\sqrt{2}$ axis

```python
#outcome1.1
import matplotlib.pyplot as plt

# Given data
data = {
        "00": 0.0712890625,
		"01": 0.4306640625,
        "10": 0.42041015625,
		"11": 0.07763671875,


}

# Extracting keys and values for plotting
states = list(data.keys())
probabilities = list(data.values())

# Creating the bar chart
plt.figure(figsize=(8, 5))
plt.bar(states, probabilities, color='blue')
plt.xlabel('Quantum States')
plt.ylabel('Probability')
plt.title('$Z$ and $(Z+X)/\sqrt{2}$')
plt.show()
```

```python
#correlation coefficient 
#E1=E(Z,(Z+X)/\sqrt{2})
E1 = (data["00"]+data["11"])-(data["01"]+data["10"])
print(E1)
```
